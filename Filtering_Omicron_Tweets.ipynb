```python
nltk.download('punkt')
nltk.download('stopwords')

STOP_WORDS = stopwords.words()

# dfx = dfx[['created_at', 'id', 'text']]

interesting_files = glob.glob("/content/drive/MyDrive/Omicron Tweet Analytics project/Filter4_res/Weekly_Club/*.csv") 
#print(len(interesting_files))
number_of_files = len(interesting_files)

for week in range(1, number_of_files+1):    # number_of_files
  path = "/content/drive/MyDrive/Omicron Tweet Analytics project/Filter4_res/Weekly_Club/Week_" + str(week) + ".csv"
  #parent_path = os.listdir("/content/drive/MyDrive/IDS 506 Healthcare Analytics/506 Project/My_Club/Week_" + str(week) + ".csv")
  df = pd.read_csv(path, engine= 'python')
  df_reduced = df[['created_at', 'id', 'text', 'user_id']]
  print(df_reduced.shape, "\n")


  def cleaning_omicron(text):
      """
      Convert to lowercase.
      Rremove URL links, special characters and punctuation.
      Tokenize and remove stop words.
      """
      text = text.lower()
      text = re.sub('[#@]omicron', 'omicron', text)
      text = re.sub('#omicorn', 'omicron', text)
      text = re.sub('#omnicron', 'omicron', text)   
      text = re.sub('#omnicorn', 'omicron', text)
      text = re.sub('omicorn', 'omicron', text)
      text = re.sub('omnicorn', 'omicron', text)
      text = re.sub('omnicron', 'omicron', text)
      return text
      
    
      
  '''  
      # removing the stop-words
      text_tokens = word_tokenize(text)
      tokens_without_sw = [
        word for word in text_tokens if not word in STOP_WORDS]
      filtered_sentence = (" ").join(tokens_without_sw)
      text = filtered_sentence
  '''
      
      
  df_reduced['text'] = df_reduced['text'].astype(str)
  df_reduced['omicron_tweets'] = df_reduced['text'].apply(cleaning_omicron)
  df_reduced = df_reduced[df_reduced['omicron_tweets'].str.contains("omicron", na=False)]
  print(df_reduced.head())
  print(df_reduced.shape)

  path2 = "/content/drive/MyDrive/Omicron Tweet Analytics project/Filter5_res/Omicron_Tweets"

  #os.makedirs(path, exist_ok=False)
  isExist = os.path.exists(path2)
  if not isExist:
  
    # Create a new directory because it does not exist 
    os.makedirs(path2)
    print("The new directory is created!")
  #dest_path = "/content/drive/MyDrive/IDS 506 Healthcare Analytics/506 Project/Clean_tweets/Clean_Week_" + rng + ".csv"
  dest_path2 = "/content/drive/MyDrive/Omicron Tweet Analytics project/Filter5_res/Omicron_Tweets/Omi_Week" + str(week) + ".csv"
  #df.to_csv(dest_path)
  df_reduced.to_csv(dest_path2, index=False)

df_reduced.head(900)

```
